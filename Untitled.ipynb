```python
import sqlite3
import os

DB_PATH = r"C:\Users\amitu\Documents\connect_now_apis\connectnow.db"

os.makedirs(os.path.dirname(DB_PATH), exist_ok=True)

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()

# =====================================================
# USER REGISTRATION
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS user_reg_info(
    unique_user_id TEXT PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT NOT NULL UNIQUE,
    city TEXT,
    gender TEXT,
    created_dt TEXT NOT NULL,
    updated_dt TEXT
)
""")

# =====================================================
# USER LOGIN
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS user_login_info(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    unique_user_id TEXT NOT NULL UNIQUE,
    email TEXT NOT NULL UNIQUE,
    password_hash TEXT NOT NULL,
    is_active INTEGER DEFAULT 1,
    created_dt TEXT NOT NULL,
    last_login_dt TEXT,

    FOREIGN KEY(unique_user_id)
    REFERENCES user_reg_info(unique_user_id)
)
""")

# =====================================================
# USER PROFILE
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS user_profile_details(

    id INTEGER PRIMARY KEY AUTOINCREMENT,

    unique_user_id TEXT NOT NULL UNIQUE,

    about_me TEXT,

    can_help_with TEXT,

    skills TEXT,

    experience TEXT,

    interests TEXT,

    availability TEXT,

    profile_completed INTEGER DEFAULT 0,

    created_dt TEXT,

    updated_dt TEXT,

    FOREIGN KEY(unique_user_id)
    REFERENCES user_reg_info(unique_user_id)

)
""")

# =====================================================
# COMMUNITY POSTS
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS community_posts(

    post_id TEXT PRIMARY KEY,

    unique_user_id TEXT NOT NULL,

    title TEXT,

    short_description TEXT,

    detailed_description TEXT,

    category TEXT,

    custom_category TEXT,

    event_date TEXT,

    event_time TEXT,

    location TEXT,

    post_type TEXT,

    status TEXT,

    created_dt TEXT,

    updated_dt TEXT,

    FOREIGN KEY(unique_user_id)
    REFERENCES user_reg_info(unique_user_id)

)
""")

# =====================================================
# CONNECTIONS
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS connections(

    connection_id TEXT PRIMARY KEY,

    requester_user_id TEXT,

    receiver_user_id TEXT,

    status TEXT,

    created_dt TEXT,

    updated_dt TEXT

)
""")

# =====================================================
# USER PRESENCE
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS user_presence(

    id INTEGER PRIMARY KEY AUTOINCREMENT,

    unique_user_id TEXT UNIQUE,

    is_online INTEGER DEFAULT 0,

    last_seen_at TEXT,

    updated_dt TEXT

)
""")

# =====================================================
# CONVERSATIONS
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS conversations(

    conversation_id TEXT PRIMARY KEY,

    user_1_id TEXT,

    user_2_id TEXT,

    created_dt TEXT,

    updated_dt TEXT,

    last_message_dt TEXT

)
""")

# =====================================================
# MESSAGES
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS messages(

    message_id TEXT PRIMARY KEY,

    conversation_id TEXT,

    sender_user_id TEXT,

    receiver_user_id TEXT,

    message_type TEXT DEFAULT 'text',

    message_text TEXT,

    attachment_url TEXT,

    reply_to_message_id TEXT,

    is_read INTEGER DEFAULT 0,

    is_deleted INTEGER DEFAULT 0,

    created_dt TEXT,

    updated_dt TEXT

)
""")

# =====================================================
# GIGS
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS gigs(

    gig_id TEXT PRIMARY KEY,

    unique_user_id TEXT,

    title TEXT,

    short_description TEXT,

    detailed_description TEXT,

    category TEXT,

    custom_category TEXT,

    budget REAL,

    location TEXT,

    deadline_date TEXT,

    status TEXT,

    created_dt TEXT,

    updated_dt TEXT

)
""")

# =====================================================
# GIG APPLICATIONS
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS gig_applications(

    application_id TEXT PRIMARY KEY,

    gig_id TEXT,

    applicant_user_id TEXT,

    message TEXT,

    status TEXT,

    created_dt TEXT,

    updated_dt TEXT

)
""")

# =====================================================
# NOTIFICATIONS
# =====================================================

cur.execute("""
CREATE TABLE IF NOT EXISTS notifications(

    notification_id TEXT PRIMARY KEY,

    unique_user_id TEXT,

    title TEXT,

    message TEXT,

    notification_type TEXT,

    is_read INTEGER DEFAULT 0,

    created_dt TEXT

)
""")

conn.commit()

print("=" * 60)
print("DATABASE CREATED SUCCESSFULLY")
print("=" * 60)
print(DB_PATH)

tables = cur.execute("""
SELECT name
FROM sqlite_master
WHERE type='table'
ORDER BY name
""").fetchall()

print("\nTables Created:\n")

for table in tables:
    print("-", table[0])

conn.close()
```

```python
import requests
import pandas as pd

OVERPASS_URL = "https://overpass-api.de/api/interpreter"

query = """
[out:json][timeout:90];

area["name"="Pune"]["boundary"="administrative"]->.searchArea;

(
  node["place"~"suburb|neighbourhood|quarter|locality"](area.searchArea);
  way["place"~"suburb|neighbourhood|quarter|locality"](area.searchArea);
  relation["place"~"suburb|neighbourhood|quarter|locality"](area.searchArea);
);

out center tags;
"""

response = requests.post(
    OVERPASS_URL,
    data={"data": query},
    headers={"User-Agent": "connectnow-locality-builder/1.0"},
    timeout=120
)

print("Status:", response.status_code)
print("Content-Type:", response.headers.get("Content-Type"))
print("First 500 chars:")
print(response.text[:500])

response.raise_for_status()

data = response.json()

rows = []

for element in data.get("elements", []):
    tags = element.get("tags", {})
    name = tags.get("name")

    if not name:
        continue

    lat = element.get("lat") or element.get("center", {}).get("lat")
    lon = element.get("lon") or element.get("center", {}).get("lon")

    rows.append({
        "city": "Pune",
        "locality": name,
        "type": tags.get("place"),
        "lat": lat,
        "lon": lon
    })

df = pd.DataFrame(rows)

if len(df) > 0:
    df = df.drop_duplicates(subset=["locality"])
    df = df.sort_values("locality").reset_index(drop=True)
```
