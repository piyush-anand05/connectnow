```python
import sqlite3
import pandas as pd

DB_PATH = r"C:\Users\amitu\Documents\connect_now_apis\connectnow.db"

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()

def add_col(table, col_def):
    try:
        cur.execute(f"ALTER TABLE {table} ADD COLUMN {col_def}")
        print("Added:", table, col_def)
    except Exception as e:
        print("Skipped:", table, col_def, "|", e)

add_col("conversations", "created_dt TEXT")
add_col("conversations", "updated_dt TEXT")
add_col("conversations", "last_message_dt TEXT")

cur.execute("""
UPDATE conversations
SET created_dt = COALESCE(created_dt, created_at),
    updated_dt = COALESCE(updated_dt, updated_at),
    last_message_dt = COALESCE(last_message_dt, last_message_at)
""")

conn.commit()

display(pd.read_sql_query("PRAGMA table_info(conversations)", conn))
display(pd.read_sql_query("SELECT * FROM conversations LIMIT 5", conn))

conn.close()
```

```python
import sqlite3
import pandas as pd

DB_PATH = r"C:\Users\amitu\Documents\connect_now_apis\connectnow.db"

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()

cur.execute("DROP TABLE IF EXISTS messages")
cur.execute("DROP TABLE IF EXISTS conversations")
cur.execute("DROP TABLE IF EXISTS user_presence")

cur.execute("""
CREATE TABLE user_presence(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    unique_user_id TEXT UNIQUE NOT NULL,
    is_online INTEGER NOT NULL DEFAULT 0,
    last_seen_at TEXT,
    updated_dt TEXT NOT NULL
)
""")

cur.execute("""
CREATE TABLE conversations(
    conversation_id TEXT PRIMARY KEY,
    user_1_id TEXT NOT NULL,
    user_2_id TEXT NOT NULL,
    created_dt TEXT NOT NULL,
    updated_dt TEXT NOT NULL,
    last_message_dt TEXT
)
""")

cur.execute("""
CREATE TABLE messages(
    message_id TEXT PRIMARY KEY,
    conversation_id TEXT NOT NULL,
    sender_user_id TEXT NOT NULL,
    receiver_user_id TEXT NOT NULL,
    message_type TEXT DEFAULT 'text',
    message_text TEXT,
    attachment_url TEXT,
    reply_to_message_id TEXT,
    is_read INTEGER DEFAULT 0,
    is_deleted INTEGER DEFAULT 0,
    created_dt TEXT NOT NULL,
    updated_dt TEXT
)
""")

conn.commit()

display(pd.read_sql_query("PRAGMA table_info(conversations)", conn))
display(pd.read_sql_query("PRAGMA table_info(messages)", conn))
display(pd.read_sql_query("PRAGMA table_info(user_presence)", conn))

conn.close()
```

```python
import sqlite3
import pandas as pd

DB_PATH = r"C:\Users\amitu\Documents\connect_now_apis\connectnow.db"

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()

# USER BLOCKS
cur.execute("""
CREATE TABLE IF NOT EXISTS user_blocks (
    block_id TEXT PRIMARY KEY,
    blocker_user_id TEXT NOT NULL,
    blocked_user_id TEXT NOT NULL,
    created_dt TEXT NOT NULL,
    UNIQUE(blocker_user_id, blocked_user_id)
)
""")

# Make sure notifications has extra reference_id column
try:
    cur.execute("ALTER TABLE notifications ADD COLUMN reference_id TEXT")
    print("Added reference_id to notifications")
except Exception as e:
    print("reference_id skipped:", e)

# Correct indexes for your current schema
indexes = [
    """
    CREATE INDEX IF NOT EXISTS idx_connections_requester
    ON connections(requester_user_id)
    """,
    """
    CREATE INDEX IF NOT EXISTS idx_connections_receiver
    ON connections(receiver_user_id)
    """,
    """
    CREATE INDEX IF NOT EXISTS idx_connections_status
    ON connections(status)
    """,
    """
    CREATE INDEX IF NOT EXISTS idx_blocks_blocker
    ON user_blocks(blocker_user_id)
    """,
    """
    CREATE INDEX IF NOT EXISTS idx_blocks_blocked
    ON user_blocks(blocked_user_id)
    """,
    """
    CREATE INDEX IF NOT EXISTS idx_notifications_user
    ON notifications(unique_user_id)
    """
]

for q in indexes:
    cur.execute(q)

conn.commit()

for table in ["connections", "user_blocks", "notifications"]:
    print("\n" + "=" * 80)
    print(table.upper())
    print("=" * 80)
    display(pd.read_sql_query(f"PRAGMA table_info({table})", conn))
    display(pd.read_sql_query(f"SELECT * FROM {table} LIMIT 5", conn))

conn.close()

print("✅ Network safety tables/indexes ready.")
```

```python
import sqlite3
import pandas as pd

from passlib.context import CryptContext

DB_PATH = r"C:\Users\amitu\Documents\connect_now_apis\connectnow.db"

email = "amitu@test.com"
new_password = "test123"

pwd_context = CryptContext(
    schemes=["bcrypt"],
    deprecated="auto"
)

new_hash = pwd_context.hash(new_password)

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()

cur.execute("""
UPDATE user_login_info
SET password_hash = ?
WHERE email = ?
""", (new_hash, email))

conn.commit()

df = pd.read_sql_query("""
SELECT
    unique_user_id,
    email,
    password_hash,
    is_active,
    created_dt,
    last_login_dt
FROM user_login_info
WHERE email = ?
""", conn, params=(email,))

display(df)

conn.close()

print("Password updated successfully.")
print("Login email:", email)
print("New password:", new_password)
```

```python
import sqlite3
import pandas as pd

DB_PATH = r"C:\Users\amitu\Documents\connect_now_apis\connectnow.db"

conn = sqlite3.connect(DB_PATH)
cur = conn.cursor()

# 1. Add post-level private reply setting
try:
    cur.execute("""
    ALTER TABLE community_posts
    ADD COLUMN allow_private_replies INTEGER DEFAULT 1
    """)
    print("Added allow_private_replies")
except Exception as e:
    print("allow_private_replies skipped:", e)

# 2. Post likes
cur.execute("""
CREATE TABLE IF NOT EXISTS post_likes (
    like_id TEXT PRIMARY KEY,
    post_id TEXT NOT NULL,
    unique_user_id TEXT NOT NULL,
    created_dt TEXT NOT NULL,
    UNIQUE(post_id, unique_user_id)
)
""")

# 3. Post reports
cur.execute("""
CREATE TABLE IF NOT EXISTS post_reports (
    report_id TEXT PRIMARY KEY,
    post_id TEXT NOT NULL,
    reporter_user_id TEXT NOT NULL,
    reason TEXT NOT NULL CHECK(reason IN ('Abuse', 'Sexual', 'Fraud', 'Other')),
    other_reason TEXT,
    status TEXT NOT NULL DEFAULT 'pending',
    created_dt TEXT NOT NULL,
    UNIQUE(post_id, reporter_user_id)
)
""")

# 4. Private reply threads
cur.execute("""
CREATE TABLE IF NOT EXISTS post_private_threads (
    thread_id TEXT PRIMARY KEY,
    post_id TEXT NOT NULL,
    post_owner_id TEXT NOT NULL,
    sender_user_id TEXT NOT NULL,
    status TEXT NOT NULL DEFAULT 'active',
    created_dt TEXT NOT NULL,
    updated_dt TEXT NOT NULL,
    last_message_dt TEXT
)
""")

# 5. Private reply messages
cur.execute("""
CREATE TABLE IF NOT EXISTS post_private_messages (
    message_id TEXT PRIMARY KEY,
    thread_id TEXT NOT NULL,
    sender_user_id TEXT NOT NULL,
    receiver_user_id TEXT NOT NULL,
    message_text TEXT NOT NULL,
    is_read INTEGER NOT NULL DEFAULT 0,
    is_deleted INTEGER NOT NULL DEFAULT 0,
    created_dt TEXT NOT NULL,
    updated_dt TEXT
)
""")

# Indexes
indexes = [
    "CREATE INDEX IF NOT EXISTS idx_post_likes_post ON post_likes(post_id)",
    "CREATE INDEX IF NOT EXISTS idx_post_likes_user ON post_likes(unique_user_id)",
    "CREATE INDEX IF NOT EXISTS idx_post_reports_post ON post_reports(post_id)",
    "CREATE INDEX IF NOT EXISTS idx_post_reports_user ON post_reports(reporter_user_id)",
    "CREATE INDEX IF NOT EXISTS idx_private_threads_post ON post_private_threads(post_id)",
    "CREATE INDEX IF NOT EXISTS idx_private_threads_owner ON post_private_threads(post_owner_id)",
    "CREATE INDEX IF NOT EXISTS idx_private_threads_sender ON post_private_threads(sender_user_id)",
    "CREATE INDEX IF NOT EXISTS idx_private_messages_thread ON post_private_messages(thread_id)"
]

for q in indexes:
    cur.execute(q)

conn.commit()

for table in [
    "community_posts",
    "post_likes",
    "post_reports",
    "post_private_threads",
    "post_private_messages"
]:
    print("\n" + "=" * 80)
    print(table.upper())
    print("=" * 80)
    display(pd.read_sql_query(f"PRAGMA table_info({table})", conn))
    display(pd.read_sql_query(f"SELECT * FROM {table} LIMIT 5", conn))

conn.close()

print("✅ Post engagement DB ready.")
```
